# Problem #1

<br>

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

<br>

Find the sum of all the multiples of 3 or 5 below 1000.

<br>

### Solution #1

```python
result = 0

for x in range(1000):
    if x % 3 == 0 or x % 5 == 0:
        result += x
print(result)
```

### Solution #2

```python
# Using list comprehension
print(sum([x for x in range(1000) if x % 3 == 0 or x % 5 == 0]))
```

### Solution #3

```python
print(sum(x for x in range(1000) if x % 3 == 0 or x % 5 == 0))
```

```python
def solution(num):
    result = 0
    for x in range(1000):
        if x % num == 0:
            result += x
    return result
```

```python
sum_3 = solution(3)
sum_5 = solution(5)
sum_15 = solution(15)
```
